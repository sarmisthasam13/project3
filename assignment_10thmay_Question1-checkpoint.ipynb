```python
kmeans=KMeans(n_clusters=1)
kmeans.fit(data1)
kmeans.inertia_
```

```python
kmeans=KMeans(n_clusters=2)
kmeans.fit(data1)
kmeans.inertia_
```

```python
labels2=kmeans.labels_
labels2
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(data1)
kmeans.inertia_
```

```python
labels3=kmeans.labels_
```

```python
wss=[]
for i in range(1,5):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data1)
    wss.append(kmeans.inertia_)
```
